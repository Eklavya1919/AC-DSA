```python
# prime number
n = int(input("Enter n: "))
for i in range(2,int(n**0.5)):
    if n%i==0:
        print("no")
        break
else:
    print("yes")
```

```python
# if-else
age = int(input("Enter age: "))
if age<18:
    print("not eligible to vote")
else:
    print("eligible to vote")
```

```python
# if else in single line
num = int(input("enter num"))
ans = "positive" if num>=0 else "negative"
print(ans)
```

```python
# character
character = input("Enter character: ")
cnum = ord(character)
if 65<=cnum<=90:
    print("uppercase")
else:
    print("lowercase")
```

```python
#  print 1 to 5
i = 1
while i<=5:
    print(i, end="")
    i+=1
print()
for i in range(1,6):
    print(i, end="")
```

```python
# sum of odd nums till n
n = int(input("Enter n: "))
fsum = 0
for i in range(1,n+1,2):
    fsum+=i
print(fsum)

wsum = 0
i=1
while i<=n:
    if i%2!=0:
        wsum+=i
        i+=1
    else:
        i+=1
print(wsum)
```

```python
# sum of numbers from 1 to n
n = int(input("Enter n: "))
fsum = 0
for i in range(n+1):
    fsum+=i
print(fsum)

wsum = 0
while n>0:
    wsum+=n
    n-=1
print(wsum)
```

## patterns using loops

```python
# square
n = int(input("Enter n: "))
for i in range(n):
    for j in range(n):
        print("* ", end="")
    print()
```

```python
# lower triangle
n = int(input("Enter n: "))
for i in range(1,n+1):
    for j in range(1,i+1):
        print("* ", end="")
    print()
```

```python
# upper triangle
n = int(input("Enter n: "))
for i in range(1,n+1):
    for j in range(n-i+1):
        print("* ", end="")
    print()
```

```python
# hollow square
n = int(input("Enter n: "))
for i in range(n):
    for j in range(n):
        if i==0 or j==0 or i==n-1 or j==n-1:
            print("* ", end="")
        else:
            print("  ", end="")
    print()
```
