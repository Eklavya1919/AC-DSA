```python
# sum of numbers from 1 to n
n = int(input("Enter n: "))
fsum = 0
for i in range(n+1):
    fsum+=i
print(fsum)

wsum = 0
while n>0:
    wsum+=n
    n-=1
print(wsum)
```

```python
# prime number
n = int(input("Enter n: "))
for i in range(2,int(n**0.5)):
    if n%i==0:
        print("no")
        break
else:
    print("yes")
```
